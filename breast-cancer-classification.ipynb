```python

x = df_cancer.drop(['target'], axis=1)
print(x)
```

```python
y=df_cancer['target']
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=5)
```

```python
x_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(x_train,y_train)
```

```python
y_predict =svc_model.predict(x_test)
```

```python
y_predict
```

```python
cm=confusion_matrix(y_test,y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_predict))
```

```python
sns.scatterplot(x = x_train['mean area'], y = x_train['mean smoothness'], hue = y_train)
```

```python
#scaling x_train
x_train_min = x_train.min()
range_train = (x_train - x_train_min).max()

x_train_scaled = (x_train - x_train_min)/ range_train
```

```python
sns.scatterplot(x = x_train_scaled['mean area'], y = x_train_scaled['mean smoothness'], hue = y_train)
```

```python
#scaling x_test
x_test_min = x_test.min()
range_test = (x_test - x_test_min).max()

x_test_scaled = (x_test - x_test_min)/ range_test
```

```python
svc_model.fit(x_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(x_test_scaled)
```

```python
cm = confusion_matrix(y_test,y_predict)
sns.heatmap(cm ,annot=True)
```

```python
print(classification_report(y_test , y_predict))
```

```python
#decide grid parameters
grid_params = {'C' : [0,1,1,10,100], 'gamma' : [1,0.1,0.01,0.001], 'kernel' : ['rbf']}
```

```python
#import grid_search_cv
from sklearn.model_selection import GridSearchCV
```

```python
#creating our own model
mymodel = GridSearchCV(SVC(), grid_params, refit=True, verbose = 4)
```

```python
#fitting the model
mymodel.fit(x_train_scaled,y_train)
```

```python
mymodel.best_params_
```

```python
y_predict = mymodel.predict(x_test_scaled)
```
